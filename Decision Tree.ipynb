```python
import pandas as pd
from chefboost import Chefboost as chef

```

# Football Example 1

```python
df = pd.read_csv("football1.csv")
df.head()
```

```python
df1 = df[df['Decision'].notnull()]
df2 = df[df['Decision'].notnull()==False]
```

## Build a tree using ID3
#tutorial:https://sefiks.com/2019/08/31/a-begineers-guide-to-decision-trees-in-python/


```python
#train
config = {'algorithm': 'ID3'}
model = chef.fit(df1[['home_away','opponent','media','Decision']], config)
```

```python
#test
for index, instance in df2[['home_away','opponent','media','Decision']].iterrows():
    prediction = chef.predict(model, instance)
    df2.at[index, 'Decision'] = [prediction]
df2
```

# Football Example 2

```python
#ID3

df1 = pd.read_csv("football2.csv")
df2 = pd.read_csv("football3.csv")

#train
config = {'algorithm': 'ID3'}
model = chef.fit(df1[['home_away','opponent','media','Decision']], config)
#test
for index, instance in df2[['home_away','opponent','media','Decision']].iterrows():
    prediction = chef.predict(model, instance)
    df2.at[index, 'Prediction'] = prediction
df2

```

```python
#test

for index, instance in df2[['home_away','opponent','media','Decision']].iterrows():
    prediction = chef.predict(model, instance)
    actual = instance['Decision']
    if actual == prediction:
        classified = True
    else:
        print('*', end = '')
    print (actual, ' - ', prediction)
    df2.at[index, 'Decision'] = prediction
```

```python
#Calculate Accuracy, Precision, Recall, and F1 score on the testing result.
accuracy = 10/12
#precision = true positive predictions / all positive predictions
precision = 7/7
# recall = correctly predicted positive observations / all true positive observations
recall = 7/9
# F1 Score = 2*(Recall * Precision) / (Recall + Precision)
F1 = 2*(recall * precision) / (recall + precision)

print ('accuracy:', accuracy, 'precision:', precision, 'recall:', recall, 'F1:', F1)
```
